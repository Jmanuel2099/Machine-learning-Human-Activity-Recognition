```python
# lib para calculos matematicos y matrices
import pandas as pd
import numpy as np
```

## Carga del dataset
Inicialmente el data set esta partido en dos partes (70% -> Entrenamiento, 30% -> Prueba)

### Carga del nombre de las columnas del dataset

```python
dataframe_features_name = pd.read_csv('./HAPT Data Set/features.txt', header=None, delimiter='\s+')
features_name = dataframe_features_name.values[:,0]
```

### Carga del dataset de prueba

```python
# se carga el data set de prueba y se asigna el header 
dataframe_xtest = pd.read_csv('./HAPT Data Set/Test/X_test.txt', delimiter=' ', header=None)
dataframe_xtest.columns = features_name

dataframe_ytest = pd.read_csv('./HAPT Data Set/Test/Y_test.txt', header=None, names=["Activity"])

dataframe_test = pd.concat([dataframe_xtest, dataframe_ytest], axis=1)
```

### Carga de dataset de entrenamiento

```python
# se carga el data set de entrenamiento y se asigna el header 
dataframe_xtrain = pd.read_csv('./HAPT Data Set/Train/X_train.txt', delimiter=' ', header=None)
dataframe_xtrain.columns = features_name

dataframe_ytrain = pd.read_csv('./HAPT Data Set/Train/Y_train.txt', header=None, names=["Activity"])

dataframe_train = pd.concat([dataframe_xtrain, dataframe_ytrain], axis=1)
```

```python
dataframe = pd.concat([dataframe_train, dataframe_test])
dataframe.head(5)
```

### Cargar activity labels

```python
dataframe_activity_labels = pd.read_csv('./HAPT Data Set/activity_labels.txt', header=None)

activity_label_dict = {}
for label in dataframe_activity_labels.values[:,0]:
    data_label = label.split(" ")
    activity_label_dict[int(data_label[0])] = data_label[1]
```

```python
activity_label_column = dataframe["Activity"].map(activity_label_dict)
dataframe["Activity-Label"] = activity_label_column

# Se crea un archivo csv con el nuevo dataframe total
dataframe.to_csv("data/dataframe_total.csv", index = False)

dataframe.head(5)
```
